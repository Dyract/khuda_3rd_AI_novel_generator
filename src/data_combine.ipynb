# 데이터 확인

```python
# 예시용 DataFrame 생성
data = {'input': ['첫 번째 텍스트입니다.', '두 번째 텍스트입니다.', '세 번째 텍스트입니다.', '네 번째 텍스트입니다.'],
        'output': [None,None, '두 번째 결과입니다.', None]}
ex_df = pd.DataFrame(data)
ex_df
```

```python
# 순회하며 'output'이 결측치인 경우 다음 행의 'input'과 이번 행의 'input'을 합친 후 다음 행의 'input'에 넣고, 'output'에는 다음 행의 'output'을 적용
for i in range(len(ex_df)):
    if pd.isnull(ex_df.loc[i, 'output']):
        next_input = ex_df.loc[i+1, 'input'] if i+1 < len(ex_df) else ''  # 다음 행의 'input'을 가져옴
        current_input = ex_df.loc[i, 'input']  # 현재 행의 'input'
        ex_df.loc[i+1, 'input'] = current_input + next_input  # 다음 행의 'input'에 합친 값을 넣음
        # ex_df.loc[i, 'output'] = ex_df.loc[i+1, 'output']  # 현재 행의 'output'에 다음 행의 'output'을 적용

# 마지막 행의 'input'은 다음 행이 없으므로 해당 행의 'output'을 None으로 처리
ex_df.loc[len(ex_df) - 1, 'output'] = None

print(ex_df)
```

```python
# 'input' 열에 결측치가 있는 행 제거
ex_df = ex_df.dropna(subset=['output'])
ex_df
```

```python
# 순회하며 'output'이 결측치인 경우 다음 행의 'input'과 이번 행의 'input'을 합친 후, 'output'은 그대로 둠
for i in range(len(df)):
    if pd.isnull(df.loc[i, 'output']):
        next_input = df.loc[i+1, 'input'] if i+1 < len(df) else ''  # 다음 행의 'input'을 가져옴
        current_input = df.loc[i, 'input']  # 현재 행의 'input'
        df.loc[i+1, 'input'] = current_input + next_input  # 다음 행의 'input'에 합친 값을 넣음

# 마지막 행의 'input'은 다음 행이 없으므로 해당 행의 'output'을 None으로 처리
df.loc[len(df) - 1, 'output'] = None

print(df)
```

```python
# 'output' 열에 결측치가 있는 행 제거
df = df.dropna(subset=['output'])
print(df)
```

```python
def remove_hanja(text):
    # 한자를 제거하는 정규표현식 패턴
    hanja_pattern = re.compile(r'[\u4e00-\u9fff]+')
    # 한자를 빈 문자열로 대체하여 제거
    result = hanja_pattern.sub('', text)
    return result
```

```python
#한자 제거
df['input'] = df['input'].apply(remove_hanja)
df['output'] = df['output'].apply(remove_hanja)
```

```python
# 모든 데이터의 길이를 'index: length' 딕셔너리 구조로 리스트에 넣기
length_dict_list = []
for column in df.columns:
    for index, data in enumerate(df['input']):
        length_dict_list.append({f'{index}': len(str(data))})

# length를 기준으로 내림차순으로 정렬
sorted_length_dict_list = sorted(length_dict_list, key=lambda x: list(x.values())[0], reverse=True)

# 결과 확인
print ('input 상위 10개')
for item in sorted_length_dict_list[:10]:
    print(item)

print('===========')
print('input 하위 10개')
for item in sorted_length_dict_list[-10:]:
    print(item)
```

```python
# 모든 데이터의 길이를 'index: length' 딕셔너리 구조로 리스트에 넣기
length_dict_list = []
for column in df.columns:
    for index, data in enumerate(df['output']):
        length_dict_list.append({f'{index}': len(str(data))})

# length를 기준으로 내림차순으로 정렬
sorted_length_dict_list = sorted(length_dict_list, key=lambda x: list(x.values())[0], reverse=True)

# 결과 확인
print ('output 상위 10개')
for item in sorted_length_dict_list[:10]:
    print(item)

print('===========')
print('output 하위 10개')
for item in sorted_length_dict_list[-10:]:
    print(item)
```

```python
# 각 원소의 세 문장만 남기고 나머지를 삭제하고, '다'로 끝나면 뒤에 온점을 추가하는 함수 정의
def keep_first_three_sentences(text):
    # 정규식을 사용하여 문장을 분리
    sentences = re.split(r'(?<=[.!?])\s*', text)
    # 처음 세 문장을 가져옴
    first_three_sentences = sentences[:3]
    # 문장 사이에 공백 추가하여 다시 결합
    cleaned_text = ' '.join(first_three_sentences)

    # 만약 문자열이 '다'로 끝나면 뒤에 온점을 추가
    if cleaned_text.endswith('다'):
        cleaned_text += '.'

    return cleaned_text

# DataFrame의 'output' 열에 적용하여 수정
df['output'] = df['output'].apply(keep_first_three_sentences)

# 결과 출력
print(df)
```

```python
# 인덱스 재정렬
df = df.reset_index(drop=True)

# 결과 출력
print(df)
```

```python
# DataFrame을 CSV 파일로 변환
df.to_csv('combined_text.csv', index=False)
```
